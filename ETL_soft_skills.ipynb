```python
import os
import pandas as pd
from langdetect import detect
import spacy
from spacy.matcher import PhraseMatcher

# Cargar el archivo CSV con las habilidades blandas
soft_skills_file = "ss_unique.csv"
soft_skills_df = pd.read_csv(soft_skills_file)

# Cargar los modelos de spaCy para español e inglés
nlp_es = spacy.load("es_core_news_sm")
nlp_en = spacy.load("en_core_web_sm")

# Función para cargar y procesar un archivo de texto
def process_text_file(file_name, folder='Applications'):
    file_path = os.path.join(folder, file_name)
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    return text

# Función para identificar las habilidades blandas en un texto usando spaCy
def identify_soft_skills(text, language):
    # Seleccionar el modelo de spaCy según el idioma
    nlp = nlp_es if language == "es" else nlp_en

    # Crear un matcher de frases para buscar habilidades blandas en el texto
    matcher = PhraseMatcher(nlp.vocab)
    
    # Añadir las habilidades blandas al matcher
    skills_phrases = [nlp(skill.lower()) for skill in soft_skills_df["Habilidad blanda"].tolist()]
    matcher.add("soft_skills", None, *skills_phrases)

    # Procesar el texto con spaCy
    doc = nlp(text)

    # Encontrar coincidencias en el texto
    matches = matcher(doc)

    # # Extraer las habilidades blandas identificadas
    # identified_skills = [doc[start:end].text.lower() for match_id, start, end in matches]

    # Extraer las habilidades blandas identificadas (con lematización)
    identified_skills = []
    for match_id, start, end in matches:
        for token in doc[start:end]:
            identified_skills.append(token.lemma_.lower())


    return identified_skills

# Procesar los archivos de texto de las postulaciones
for i in range(1, 51):
    file_number = str(i).zfill(2)
    
    # Verificar el idioma del texto
    text = process_text_file(f"App_{file_number}.txt", folder="Applications")
    language = detect(text)

    # Identificar habilidades blandas
    skills = identify_soft_skills(text, language)

    # Imprimir resultados
    print(f"\nPostulación {file_number}:")
    print(f"Idioma: {language}")
    print(f"Habilidades blandas identificadas: {skills}")
```

```python
import os
import pandas as pd
from langdetect import detect
import spacy

# Cargar el archivo CSV con las habilidades blandas
soft_skills_file = "ss_unique.csv"
soft_skills_df = pd.read_csv(soft_skills_file)

# Cargar los modelos de spaCy para español e inglés
nlp_es = spacy.load("es_core_news_sm")
nlp_en = spacy.load("en_core_web_sm")

# Función para cargar y procesar un archivo de texto
def process_text_file(file_name, folder='Applications'):
    file_path = os.path.join(folder, file_name)
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    return text

# Función para identificar las habilidades blandas en un texto usando spaCy
def identify_soft_skills(text, language):
    # Seleccionar el modelo de spaCy según el idioma
    nlp = nlp_es if language == "es" else nlp_en

    # Procesar el texto con spaCy
    doc = nlp(text)

    # Crear una lista de habilidades blandas lematizadas
    soft_skills_lemmas = []
    for skill in soft_skills_df["Habilidad blanda"].tolist():
        skill_doc = nlp(skill.lower())
        lemmas = [token.lemma_ for token in skill_doc]
        soft_skills_lemmas.append(" ".join(lemmas))

    # Identificar habilidades blandas en el texto lematizado
    identified_skills = []
    for skill in soft_skills_lemmas:
        if skill.lower() in text.lower():
            identified_skills.append(skill)

    return identified_skills

# Procesar los archivos de texto de las postulaciones
for i in range(1, 51):
    file_number = str(i).zfill(2)
    
    # Verificar el idioma del texto
    text = process_text_file(f"App_{file_number}.txt", folder="Applications")
    language = detect(text)

    # Identificar habilidades blandas
    skills = identify_soft_skills(text, language)

    # Imprimir resultados
    print(f"\nPostulación {file_number}:")
    print(f"Idioma: {language}")
    print(f"Habilidades blandas identificadas: {skills}")
```

```python
import os
import pandas as pd
from langdetect import detect
import spacy
from nltk import word_tokenize
from nltk.stem import WordNetLemmatizer

# Cargar el archivo CSV con las habilidades blandas
soft_skills_file = "ss_unique.csv"
soft_skills_df = pd.read_csv(soft_skills_file)

# Cargar los modelos de spaCy para español e inglés
nlp_es = spacy.load("es_core_news_sm")
nlp_en = spacy.load("en_core_web_sm")

# Inicializar lematizador de nltk para inglés
lemmatizer_en = WordNetLemmatizer()

# Función para cargar y procesar un archivo de texto
def process_text_file(file_name, folder='Applications'):
    file_path = os.path.join(folder, file_name)
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    return text

# Función para lematizar una lista de palabras en inglés
def lemmatize_english(words):
    return [lemmatizer_en.lemmatize(word.lower()) for word in words]

# Función para identificar las habilidades blandas en un texto usando spaCy y nltk
def identify_soft_skills(text, language):
    # Seleccionar el modelo de spaCy según el idioma
    nlp = nlp_es if language == "es" else nlp_en

    # Procesar el texto con spaCy
    doc = nlp(text)

    # Crear una lista de habilidades blandas lematizadas para ambos idiomas
    soft_skills_lemmas = []
    for skill in soft_skills_df["Habilidad blanda"].tolist():
        if language == 'es':
            skill_doc = nlp(skill.lower())
            lemmas = [token.lemma_ for token in skill_doc]
            soft_skills_lemmas.append(" ".join(lemmas))
        elif language == 'en':
            # Utilizar nltk para lematizar en inglés
            skill_tokens = word_tokenize(skill.lower())
            lemmas = lemmatize_english(skill_tokens)
            soft_skills_lemmas.append(" ".join(lemmas))

    # Identificar habilidades blandas en el texto lematizado
    identified_skills = []
    for skill in soft_skills_lemmas:
        if skill.lower() in text.lower():
            identified_skills.append(skill)

    return identified_skills

# Procesar los archivos de texto de las postulaciones
for i in range(1, 51):
    file_number = str(i).zfill(2)
    
    # Verificar el idioma del texto
    text = process_text_file(f"App_{file_number}.txt", folder="Applications")
    language = detect(text)

    # Identificar habilidades blandas
    skills = identify_soft_skills(text, language)

    # Imprimir resultados
    print(f"\nPostulación {file_number}:")
    print(f"Idioma: {language}")
    print(f"Habilidades blandas identificadas: {skills}")
```

```python
import os
import pandas as pd
from langdetect import detect
from textblob import TextBlob
import spacy

# Cargar el archivo CSV con las habilidades blandas
soft_skills_file = "ss_unique.csv"
soft_skills_df = pd.read_csv(soft_skills_file)

# Cargar los modelos de spaCy para español e inglés
nlp_es = spacy.load("es_core_news_sm")
nlp_en = spacy.load("en_core_web_sm")

# Función para cargar y procesar un archivo de texto
def process_text_file(file_name, folder='Applications'):
    file_path = os.path.join(folder, file_name)
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    return text

# Función para lematizar una lista de palabras en inglés usando textblob
def lemmatize_english(words):
    return [word.lemmatize().lower() for word in words]

# Función para identificar las habilidades blandas en un texto usando spaCy y textblob
def identify_soft_skills(text, language):
    # Seleccionar el modelo de spaCy según el idioma
    nlp = nlp_es if language == "es" else nlp_en

    # Procesar el texto con spaCy
    doc = nlp(text)

    # Crear una lista de habilidades blandas lematizadas para ambos idiomas
    soft_skills_lemmas = []
    column_name = "Habilidad blanda" if language == 'es' else 'Soft skill'
    
    for skill in soft_skills_df[column_name].tolist():
        if language == 'es':
            skill_doc = nlp(skill.lower())
            lemmas = [token.lemma_ for token in skill_doc]
            soft_skills_lemmas.append(" ".join(lemmas))
        elif language == 'en':
            # Utilizar textblob para lematizar en inglés
            skill_blob = TextBlob(skill.lower())
            lemmas = lemmatize_english(skill_blob.words)
            soft_skills_lemmas.append(" ".join(lemmas))

    # Identificar habilidades blandas en el texto lematizado
    identified_skills = []
    for skill in soft_skills_lemmas:
        if skill.lower() in text.lower():
            identified_skills.append(skill)

    return identified_skills

# Procesar los archivos de texto de las postulaciones
for i in range(1, 51):
    file_number = str(i).zfill(2)
    
    # Verificar el idioma del texto
    text = process_text_file(f"App_{file_number}.txt", folder="Applications")
    language = detect(text)

    # Identificar habilidades blandas
    skills = identify_soft_skills(text, language)

    # Imprimir resultados
    print(f"\nPostulación {file_number}:")
    print(f"Idioma: {language}")
    print(f"Habilidades blandas identificadas: {skills}")
```

```python
import os
import pandas as pd
from langdetect import detect
import spacy

# Cargar el archivo CSV con las habilidades blandas
soft_skills_file = "ss_unique.csv"
soft_skills_df = pd.read_csv(soft_skills_file)

# Cargar el modelo de spaCy para inglés
nlp_en = spacy.load("en_core_web_sm")

# Función para cargar y procesar un archivo de texto
def process_text_file(file_name, folder='Applications'):
    file_path = os.path.join(folder, file_name)
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    return text

# Función para identificar las habilidades blandas en un texto en inglés
def identify_soft_skills_en(text):
    identified_skills = set()
    
    # Procesar el texto con spaCy
    doc = nlp_en(text)

    # Buscar coincidencias en habilidades blandas
    for _, row in soft_skills_df.iterrows():
        skill_en = str(row['Soft skill']).lower()
        if skill_en in text.lower():
            identified_skills.add(skill_en)

    return list(identified_skills)

# Procesar los archivos de texto de las postulaciones
for i in range(1, 51):
    file_number = str(i).zfill(2)
    
    # Verificar el idioma del texto
    text = process_text_file(f"App_{file_number}.txt", folder="Applications")
    language = detect(text)

    # Identificar habilidades blandas en inglés
    skills_en = identify_soft_skills_en(text)

    # Imprimir resultados
    print(f"\nPostulación {file_number}:")
    print(f"Idioma: {language}")
    print(f"Habilidades blandas identificadas (en inglés): {skills_en}")
```
